### Import

### Reading File

```python
data = pd.read_csv('../data/jobs_bayt.csv')
```

#### Date Posted column

```python
data['Date Posted'] = data['Date Posted'].apply(lambda x: x.replace('Date Posted: ','').replace(' ','/') )
```

```python
data['Date Posted'].isnull().sum()
```

```python
def to_datetime(text_date):
    date = pd.to_datetime(text_date + " 2020").date()
    date2019 = pd.to_datetime(text_date + " 2019").date()
    if date > datetime.today().date():
        
        return date2019
    else :
        return date
data['Date Posted'] = data['Date Posted'].map(lambda x: to_datetime(x) 
                                              if x != 'Feb/29'
                                              else to_datetime('Mar/1'))
```

#### Job Description column

```python
data['Job Description'] =data['Job Description'].apply(lambda x: regex.sub(r'\n', ' ', x.replace('\xa0', '').replace('Job Description','').replace('\t','')) )
```

```python
data['Job Description'] 
```

#### Job Location column

```python
data['Job Location']
```

```python
data['Job Location'].fillna("",inplace = True)
data['Job Country'] =data['Job Location'].apply(lambda x: str(x).split(',')[-1].strip() if len(x.split(',')) > 1 else x )
data['Job Country']
```

```python

data['Job Location'].fillna("",inplace = True)
data['Job City'] =data['Job Location'].apply(lambda x: str(x).split(',')[-2].strip() if len(x.split(',')) > 1 else 'NaN' )
data['Job City']
```

#### Years of Experience column

```python
data['Years of Experience']
```

```python
data['Years of Experience'].fillna(np.nan,inplace = True)
data['Min Years of Experience'] = data['Years of Experience'].apply(lambda x: x.split('Max:')[0].replace('Min: ','')
                                                                    if len(str(x).split('Max:')) > 1
                                                                    else str(x).replace('Min: ','') )
```

```python
# if there is nan in years column
data['Min Years of Experience'] = data['Min Years of Experience'].apply(lambda x: np.nan if x =='nan' else x)
# if there is only max value in year column
data['Min Years of Experience'] = data['Min Years of Experience'].apply(lambda x: np.nan if x ==' ' else x)
#data['Min Years of Experience'] = data['Min Years of Experience'].astype(float)
```

```python
data['Max Years of Experience'] = data['Years of Experience'].apply(lambda x: x.split('Max:')[1]
                                                                    if len(str(x).split('Max:')) > 1
                                                                    else np.nan )
#data['Max Years of Experience'] = data['Max Years of Experience'].astype(float)
```

```python
data['Min Years of Experience'].unique()
```

```python
data.drop(columns='Years of Experience', inplace =True)
```

#### Age column

```python
data['Age']
```

```python
data['Age'].fillna(np.nan,inplace = True)
data['Min Age'] = data['Age'].apply(lambda x: x.split('Max:')[0].replace('Min: ','')
                                    if len(str(x).split('Max:')) > 1
                                    else str(x).replace('Min: ','') )
```

```python
# if there is nan in Age column
data['Min Age'] = data['Min Age'].apply(lambda x: np.nan if x =='nan' else x)
# if there is only max value in Age column
data['Min Age'] = data['Min Age'].apply(lambda x: np.nan if x ==' ' else x)
#data['Min Age'] = data['Min Age'].astype(float)
```

```python
data['Max Age'] = data['Age'].apply(lambda x: x.split('Max:')[1]
                                    if len(str(x).split('Max:')) > 1
                                    else np.nan )
#data['Max Age'] = data['Max Age'].astype(float)
```

```python
data.drop(columns='Age', inplace =True)
```

#### Monthly Salary Range column

```python
data['Monthly Salary Range'].unique()
```

```python
data ['Monthly Salary Min Range'] = data ['Monthly Salary Range'].apply(lambda x: x.split(' - ')[0] 
                                                                        if x != 'Unspecified'
                                                                        else  x )
```

```python
data ['Monthly Salary Max Range'] = data ['Monthly Salary Range'].apply(lambda x: x.split(' - ')[1] 
                                                                        if x != 'Unspecified'
                                                                        else  x )
```

```python
data ['Monthly Salary Min Range'] = data ['Monthly Salary Min Range'].apply(lambda x: x.replace('Unspecified','')
                                                                           .replace('$','').replace(',',''))
data ['Monthly Salary Min Range'] = data ['Monthly Salary Min Range'].apply(lambda x: np.nan if x =='' else x )
#data ['Monthly Salary Min Range'] = data ['Monthly Salary Min Range'].apply(lambda x: np.nan if x =='' else float(x) )

data ['Monthly Salary Max Range'] = data ['Monthly Salary Max Range'].apply(lambda x: x.replace('Unspecified','')
                                                                           .replace('$','').replace(',',''))
data ['Monthly Salary Max Range'] = data ['Monthly Salary Max Range'].apply(lambda x: np.nan if x =='' else x )
#data ['Monthly Salary Max Range'] = data ['Monthly Salary Max Range'].apply(lambda x: np.nan if x =='' else float(x) )
```

```python
data['Monthly Salary Min Range'].unique()
```

```python
data['Monthly Salary Max Range'].unique()
```

```python
data.drop(columns='Monthly Salary Range', inplace =True)
```

#### Employment Type

```python
data['Employment Type'].unique()
```

```python
data['Employment Type'] = data['Employment Type'].apply(lambda x: x.replace('--','Unspecified'))
```
